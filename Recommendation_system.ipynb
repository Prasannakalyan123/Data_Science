```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
book=pd.read_csv("book.csv")
book
```

```python
book=book.rename({"User.ID":"id","Book.Title":"title","Book.Rating":"rating"},axis=1)
```

```python
book
```

```python
book.id.nunique()
```

```python
book.title.nunique()
```

```python
book.rating.value_counts()
```

```python
book_df = book.pivot(index='title',
                                 columns='Unnamed: 0',
                                 values='rating').fillna(0)
```

```python
book_df
```

```python
from scipy.sparse import csr_matrix

book_df_matrix = csr_matrix(book_df.values)

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(book_df_matrix)
```

```python
book_df_matrix.shape
```

```python
query_index = np.random.choice(book_df.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(book_df.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
book_df.head()
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(book_df.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, book_df.index[indices.flatten()[i]], distances.flatten()[i]))
```
